```python
def convert_feet_to_metre(df):
    """
    Convert all feet values in a pandas DataFrame to metre values.
    """
    feet_to_metre = 0.3048
    columns_to_convert = ['v_length', 'Space_Headway', 'v_Vel', 'v_Acc', 'Local_X', 'Local_Y']
    df[columns_to_convert] = df[columns_to_convert].apply(lambda x: x * feet_to_metre)
    return df

df = convert_feet_to_metre(df)
print(df[['v_length' , 'Space_Headway' , 'v_Vel' , 'v_Acc', 'Local_X' , 'Local_Y']])
```

```python
## filtering the highway locations:

def Filter_highways(df):
    """
    Filter a DataFrame to create two separate DataFrames for US-101 and I-80 highways.
    """
    highway_u_mask = df['Location'] == 'us-101'
    highway_i_mask = df['Location'] == 'i-80'
    highway_u_df = df[highway_u_mask].copy()
    highway_i_df = df[highway_i_mask].copy()
    return highway_u_df, highway_i_df

Highway_U, Highway_I = Filter_highways(df)
print(Highway_U , Highway_I)
```

```python
def filter_vehicle_classes(df):
    """
    Filter a DataFrame to create sets of vehicle IDs for each vehicle class (motorcycle, car, and heavy vehicle).
    """
    vehicle_classes = df[['Vehicle_ID', 'v_Class', 'v_length']].drop_duplicates().sort_values('v_Class')
    m_class = set(vehicle_classes[vehicle_classes['v_Class'] == 1]['Vehicle_ID'])
    c_class = set(vehicle_classes[vehicle_classes['v_Class'] == 2]['Vehicle_ID'])
    hv_class = set(vehicle_classes[vehicle_classes['v_Class'] == 3]['Vehicle_ID'])
    return m_class, c_class, hv_class

m_u_class, c_u_class, hv_u_class = filter_vehicle_classes(Highway_U)
m_i_class, c_i_class, hv_i_class = filter_vehicle_classes(Highway_I)
```

```python
##Find and populate the preceding Vehicle Class Name into Preceding_Vehicle_Class column, and populate the Vehicle_combination as well.

def preceding_vehicle(df, m_Class, c_Class, hv_Class):
    df['Preceding_Vehicle'] = df['Preceding'].map(
        lambda x: 'Car' if x in c_Class else
                  'Heavy Vehicle' if x in hv_Class else
                  'Motorcycle' if x in m_Class else
                  'Free Flow'
    )
    df['Vehicle_combination'] = df['Preceding_Vehicle'] + '-' + df['v_Class_Name']
    return df

Highway_U = preceding_vehicle(Highway_U, m_u_class, c_u_class, hv_u_class)
Highway_I = preceding_vehicle(Highway_I, m_i_class, c_i_class, hv_i_class)
```

```python
def preced_vehicle_length(df):
    motor_lengths = df[['Vehicle_ID', 'v_length']].drop_duplicates()
    length = motor_lengths.groupby(['Vehicle_ID']).mean().to_dict()['v_length']
    df["preced_vehicle_length"] = df["Preceding"].map(length).fillna(0)
    return df

Highway_U = preced_vehicle_length(Highway_U)
Highway_I = preced_vehicle_length(Highway_I)

```

```python
##Change the Front to Front Space Headway to Back Bumper (Lead Vehicle) to Front Bumper (Following Vehicle) Headway. 
##This will return the real distance between the vehicles while disregarding vehicle length. 
##Determine how long it will take the following car to reach the back bumper of the preceding vehicle.

def change_front_to_back(df):
    df['Rear_to_Front_Space_Headway'] = df['Space_Headway'] - df['preced_vehicle_length']
    df['Front_To_Rear_Time_Headway'] = df['Rear_to_Front_Space_Headway'] / df['v_Vel']
    df['Front_To_Rear_Time_Headway'].fillna(9999, inplace=True)
    df['Front_To_Rear_Time_Headway'].replace(np.inf, 9999, inplace=True)
    return df

Highway_U = change_front_to_back(Highway_U)
Highway_I = change_front_to_back(Highway_I)
print(Highway_U, Highway_I)
```

```python
def transforming_map_lead_Vehicle(df):
    Lane_check = df[['Vehicle_ID','Lane_ID']]
    Lane_check.drop_duplicates(inplace=True)
    Lane_check = Lane_check.groupby(['Vehicle_ID'], as_index=False).count()
    
    change_lane_vehicles = set(Lane_check[Lane_check["Lane_ID"] > 1]['Vehicle_ID'])
    
    df['lane_changes'] = df['Vehicle_ID'].isin(change_lane_vehicles)
    df_2 = df[['Vehicle_ID', 'Global_Time','v_Vel', 'v_Acc', 'lane_changes', 'Local_Y', 'v_Class']]
    df_2.rename(columns={'Vehicle_ID': 'Lead_Vehicle_ID', 'v_Vel': 'Lead_Vehicle_Vel','v_Acc': 'Lead_Vehicle_Acc', "lane_changes": "Lead_lane_change", "Local_Y": "Lead_Local_Y", "v_Class": "Lead_v_Class"}, inplace=True)
    print (df_2.columns)
    df['Lead_Vehicle_ID'] = df['Preceding']
    df = df.merge(right=df_2, how='left', on=('Lead_Vehicle_ID', 'Global_Time'))

    #df = pd.concat([df.set_index(['Lead_Vehicle_ID', 'Global_Time']), df_2.set_index(['Lead_Vehicle_ID', 'Global_Time'])], axis=1, join='outer').reset_index()
    
    df['Lead_Vehicle_Vel'] = df['Lead_Vehicle_Vel'].fillna(0)
    
    df['Lead_Vehicle_Acc'] = df['Lead_Vehicle_Acc'].fillna(0)
    
    df['Lead_Local_Y'] = df['Lead_Local_Y'].fillna(9999)
    

    df['Lead_lane_change'] = df['Lead_lane_change'].fillna(False)
    
    df['Lead_v_Class'] = df['Lead_v_Class'].fillna(0)
    df['Diff_Vel_follow_lead'] = df['v_Vel'] - \
        df['Lead_Vehicle_Vel']
    df['Diff_Acc_follow-lead'] = df['v_Acc'] - \
        df['Lead_Vehicle_Acc']
    df["old_Acc"] = df.groupby(["lead-Front_Pair"], as_index=False)["v_Acc"].shift(+1)
        
    df["jerk"] = (df["v_Acc"] - df["old_Acc"]) / 0.1

        
    return df

Highway_U = transforming_map_lead_Vehicle(Highway_U)
Highway_I = transforming_map_lead_Vehicle(Highway_I)
```

```python
def pair_of_time(df):
    df = df.sort_values(by='Global_Time', ascending=True, ignore_index=True)
    df['pair_Time_Duration'] = (df.groupby('lead-Front_Pair').cumcount() * 0.1)
    
    
    max_durations = df.groupby('lead-Front_Pair')['pair_Time_Duration'].max().to_dict()
    df['total_pair_duration'] = df['lead-Front_Pair'].map(max_durations)
    
    return df
    
Highway_U=pair_of_time(Highway_U)
Highway_I=pair_of_time(Highway_I)
```

```python
def lane_change_info(df):
    # Get preceding and following lane change data
    preceding_data = tuple(preceding_lane_change_data(df))
    following_data = tuple(following_lane_change_data(df))
    
    # Concatenate dataframes
    dfs = [preceding_data[i].append(following_data[i]) for i in range(3)]
    
    # Return concatenated dataframes
    return tuple(dfs)
```

```python
def preceding_lane_change_data(df):
        lane_verify = df[['Vehicle_ID', 'Lane_ID', 'Preceding']]
        #print(f"{lane_verify.duplicated().sum()} duplicate values have been removed")
        lane_verify.drop_duplicates(inplace=True)
        lane_verify.sort_values(['Vehicle_ID', 'Lane_ID'])
        lane_verify_1 = lane_verify[((lane_verify['Lane_ID'] == 1))].drop(
            columns='Lane_ID', axis=1)
        lane_verify_2 = lane_verify[((lane_verify['Lane_ID'] == 2))].drop(
            columns='Lane_ID', axis=1)
        lane_verify_3 = lane_verify[((lane_verify['Lane_ID'] == 3))].drop(
            columns='Lane_ID', axis=1)

        lf_pair_remove_first_5_seconds_lane1 = find_preceding_lane_change_pairs(
            lane_verify_1)
        lf_pair_remove_first_5_seconds_lane2 = find_preceding_lane_change_pairs(
            lane_verify_2)
        lf_pair_remove_first_5_seconds_lane3 = find_preceding_lane_change_pairs(
            lane_verify_3)
        return lf_pair_remove_first_5_seconds_lane1,    lf_pair_remove_first_5_seconds_lane2,    lf_pair_remove_first_5_seconds_lane3
```

```python
 def following_lane_change_data(df):
        lane_verify = df[['Vehicle_ID', 'Lane_ID', 'Following']]
        #print(f"{lane_verify.duplicated().sum()} duplicate values have been removed")
        lane_verify.drop_duplicates(inplace=True)
        lane_verify.sort_values(['Vehicle_ID', 'Lane_ID'])
        lane_verify_1 = lane_verify[((lane_verify['Lane_ID'] == 1))].drop(
            columns='Lane_ID', axis=1)
        lane_verify_2 = lane_verify[((lane_verify['Lane_ID'] == 2))].drop(
            columns='Lane_ID', axis=1)
        lane_verify_3 = lane_verify[((lane_verify['Lane_ID'] == 3))].drop(
            columns='Lane_ID', axis=1)

        lf_pair_remove_first_5_seconds_lane1 = find_following_lane_change_pairs(
            lane_verify_1)
        lf_pair_remove_first_5_seconds_lane2 = find_following_lane_change_pairs(
            lane_verify_2)
        lf_pair_remove_first_5_seconds_lane3 = find_following_lane_change_pairs(
            lane_verify_3)
        return lf_pair_remove_first_5_seconds_lane1,    lf_pair_remove_first_5_seconds_lane2,    lf_pair_remove_first_5_seconds_lane3
```

```python
def find_following_lane_change_pairs(lane_verify_df):
    lane_change_vehicles = lane_verify_df.sort_values(['Vehicle_ID']).groupby('Vehicle_ID').count()

    vehicles_with_preceding_change_lane = lane_change_vehicles[lane_change_vehicles['Following'] > 1].index.values

    df_reqd = lane_verify_df[lane_verify_df['Vehicle_ID'].isin(vehicles_with_preceding_change_lane)]
    df_reqd['lf_pair'] = df_reqd['Vehicle_ID'].astype(str) + '-' + df_reqd['Following'].astype(str)
    lf_pair_remove_first_5_seconds_lane = df_reqd['lf_pair']
    return lf_pair_remove_first_5_seconds_lane

def find_preceding_lane_change_pairs(lane_verify_df):
    lane_change_vehicles = lane_verify_df.sort_values(['Vehicle_ID']).groupby('Vehicle_ID').count()

    vehicles_with_preceding_change_lane = lane_change_vehicles[lane_change_vehicles['Preceding'] > 1].index.values

    df_reqd = lane_verify_df[lane_verify_df['Vehicle_ID'].isin(vehicles_with_preceding_change_lane)]
    df_reqd['lf_pair'] = df_reqd['Preceding'].astype(str) + '-' + df_reqd['Vehicle_ID'].astype(str)
    lf_pair_remove_first_5_seconds_lane = df_reqd['lf_pair']
    return lf_pair_remove_first_5_seconds_lane
```

```python
def filter_records_for_model(df):
    
    loc = df['Location'][0]

    # Remove vehicles with multiple lengths and classes for the same vehicle ID
    v_class_counts = df[['Lead_Vehicle_ID', 'Lead_v_Class']].drop_duplicates()\
                                                            .groupby('Lead_Vehicle_ID')\
                                                            .size()
    remove_bad_data_vehicles = set(v_class_counts[v_class_counts > 1].index)
    bad_v_class_records = df[(df['Lead_Vehicle_ID'].isin(remove_bad_data_vehicles)) | 
                             (df['Preceding'].isin(remove_bad_data_vehicles))]

    # Remove vehicles with total car following time less than 30 seconds
    total_pair_duration_mask = (df['total_pair_duration'] < 30)
    total_duration_less_than_minute = df[total_pair_duration_mask]

    # Remove vehicles with time headway less than 5 seconds
    time_headway_mask = (df['Time_Headway'] < 5)
    time_headway_less_than5 = df[time_headway_mask]

    # Remove vehicles on ramps
    remove_ramp_data = df[(df['Lane_ID'] >= 4)]

    # Remove vehicles that change lanes within the first and last 5 seconds
    lane_changes = lane_change_info(df)
    both_lane_change_dfs = [df[(df['lead-Front_Pair'].isin(lane_changes[i])) & 
                               (df['Lane_ID'] == i+1) &
                               ((df['pair_Time_Duration'] < 5) | 
                                (df['pair_Time_Duration'] > (df['total_pair_duration'] - 5)))]
                            for i in range(3)]
    
    # Concatenate all dataframes to be removed
    remove = pd.concat([time_headway_less_than5, 
                        total_duration_less_than_minute, 
                        bad_v_class_records] + both_lane_change_dfs + [remove_ramp_data])

    # Drop rows to be removed from the dataframe
    before = len(df)
    df = df.drop(remove.index)
    after = len(df)
    removed_row_count = before - after

    # Print summary of removed rows
    print(f"{loc}: {removed_row_count} rows removed using above criteria")
    for i, (lane_df, lane_name) in enumerate(zip(both_lane_change_dfs, ['Lane 1', 'Lane 2', 'Lane 3'])):
        print(f"{loc}: {lane_df.shape[0]} vehicles first and last 5 seconds removed from {lane_name} due to lane changing")
    print(f"{loc}: {bad_v_class_records.shape[0]} have multiple lengths and classes for the same vehicle ID")
    print(f"{loc}: {total_duration_less_than_minute.shape[0]} have total car following less than 30 seconds")
    print(f"{loc}: {time_headway_less_than5.shape[0]} have time headway less than 5 seconds")
    print(f"{loc}: {remove_ramp_data.shape[0]} have Lane ID as 7 or 8 which are the ramps")

    return df
```

```python
 def train_test_pair(df, split, neural=False, seed=0):
        '''
        Read the input file into a dataframe.
        Input: File name for the file present in Data folder.
        Output: Dataframe name.
        '''
        import random
        if seed > 0:
            random.seed(seed)

        total_pairs = df["lead-Front_Pair"].unique()
        total_pairs = total_pairs.tolist()
        test_split_cnt = round(len(total_pairs)*split)
        test_split_pairs = random.sample(total_pairs, test_split_cnt)
        train_df = df[df['lead-Front_Pair'].isin(test_split_pairs)]
        test_df = df[~df['lead-Front_Pair'].isin(test_split_pairs)]
        if neural:
            validation_split_cnt = round(test_split_cnt*0.1)
        else:
            validation_split_cnt = round(test_split_cnt*0.0)

        validation_split_pairs = random.sample(
            test_split_pairs, validation_split_cnt)
        val_df = df[df['lead-Front_Pair'].isin(validation_split_pairs)]
        train_df = df[~df['lead-Front_Pair'].isin(validation_split_pairs)]
        return train_df, val_df, test_df
```
